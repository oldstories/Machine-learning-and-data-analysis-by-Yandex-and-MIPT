```python
# импорт необходимых библиотек
import numpy as np
import math
import pylab as pl
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
b = 3 # параметр функции Парето, отвечающий за "гладкость"
pareto_rv = sts.pareto(b) # определяем функцию Парето
sample = pareto_rv.rvs(1000) #генерируем выборку из 1000 случ. значений
x = np.linspace(1,7,1000) # создаем массив точек для выборки
pdfx = pareto_rv.pdf(x) #вычисляем PDF (плотность распределения)
```

```python
#строим гистограмму распределения значений выборки и кривую плотности распределения
pl.title ('Гистограмма распределения Парето и его PDF') # заголовок
pl.xlabel('x') # подпись соответствующей оси
pl.ylabel('number of f(x)') # подпись соответствующей оси
pl.hist(sample,bins = 40, range =(1,3), normed = True) # Гистограмма распределения значений выборки
plt.plot(x, pdfx, label='theoretical PDF', color = 'red') # кривая плотности распределения значений
pl.legend (loc = 'upper right') # положение легенды
```

```python
s1=np.array([]) # массив для выборки1 
n=7 # задаем значение n
i=1 # начальное значение счетчика
while i <= 1000:# генерируем выборку циклом с предусловием
    si1 = pareto.rvs(n)#генерация выборки распределения Парето
    mi1=sum(si1)/n #вычисление выборочного среднего
    s1=np.append(s1,mi1)#наполнение массива
    i+=1 #увеличение счетчика
```

```python
m=1.5 #среднее распределения Парето
dis=6.75 #его дисперсия 
dis1=dis/n #дисперсия для нормального распределения выборки средних
sigma1=math.sqrt(dis1)#сигма для нормального распределения выборки средних
norm1 = sts.norm(m, sigma1)#генерируем нормальное распределение Парето
pdf1 = norm1.pdf(x[:500])# плотность распределения
plt.xlabel('x') 
plt.ylabel('number of samples') 
plt.hist(s1, bins =40, range=(1,3), normed=True)
plt.plot(x[:500], pdf1, label='PDFдля n=7', color = 'red')
pl.legend (loc = 'upper right')
```

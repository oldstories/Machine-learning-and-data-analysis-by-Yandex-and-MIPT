Импортируем все необходимые библиотеки:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import chi2
%matplotlib inline
```

Сгенерируем выборку объёма 100 из  распределения chi^2 (с k степенями свободы = df) и построим гистограмму:

```python
df = 10
sample_1 = chi2.rvs(df, loc=0, size=1000)

plt.hist(sample_1, normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

# на том же графике построим теоретическую плотность распределения:
x = np.linspace(0,30,1000)
pdf = chi2.pdf(x, 10)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

При трёх и более значениях n (5, 10, 50) сгенерируем 1000 выборок объёма n и построим гистограммы распределений их выборочных средних:

```python
sample_1 = []
for i in range(1000):
    sample = chi2.rvs(df, loc=0, size=5)
    sample_1.append(sample.mean()) 

plt.hist(sample_1, normed = True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
sample_2 = []
for i in range(1000):
    sample = chi2.rvs(df, loc=0, size=10)
    sample_2.append(sample.mean()) 

plt.hist(sample_2, normed = True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
sample_3 = []
for i in range(1000):
    sample = chi2.rvs(df, loc=0, size=50)
    sample_3.append(sample.mean()) 

plt.hist(sample_3, normed = True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
from statsmodels.stats.weightstats import _tconfint_generic
```

```python
data_0_mean = data_0.mean()
data_1_mean = data_1.mean()
data_0_mean_std = data_0.std(ddof=1)/sqrt(len(data_0))
data_1_mean_std = data_1.std(ddof=1)/sqrt(len(data_1))
```

```python
print ("data_0 mean 95%% confidence interval", _tconfint_generic(data_0_mean, data_0_mean_std,
                                                                       len(data_0) - 1,
                                                                       0.05, 'two-sided'))

print ("data_1 mean 95%% confidence interval", _tconfint_generic(data_1_mean, data_1_mean_std,
                                                                         len(data_1) - 1,
                                                                         0.05, 'two-sided'))
```

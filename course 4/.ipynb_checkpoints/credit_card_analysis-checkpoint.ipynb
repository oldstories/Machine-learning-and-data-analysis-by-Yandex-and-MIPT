```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats
```

```python
data = pd.read_csv('credit_card_default_analysis.csv')
```

```python
data
```

```python
data_1 = data.LIMIT_BAL[data.default == 1]
data_0 = data.LIMIT_BAL[data.default == 0]
```

```python
hist(data_0)
hist(data_1)
```

a) о равенстве медианных значений кредитного лимита с помощью подходящей интервальной оценки 

```python
from statsmodels.stats.weightstats import _tconfint_generic
```

```python
data_0_mean = data_0.mean()
data_1_mean = data_1.mean()
data_0_mean_std = data_0.std(ddof=1)/sqrt(len(data_0))
data_1_mean_std = data_1.std(ddof=1)/sqrt(len(data_1))
```

```python
print ("data_0 mean 95%% confidence interval", _tconfint_generic(data_0_mean, data_0_mean_std,
                                                                       len(data_0) - 1,
                                                                       0.05, 'two-sided'))

print ("data_1 mean 95%% confidence interval", _tconfint_generic(data_1_mean, data_1_mean_std,
                                                                         len(data_1) - 1,
                                                                         0.05, 'two-sided'))
```

b) о равенстве распределений с помощью одного из подходящих непараметрических критериев проверки равенства средних

$H_0\colon$ Распределения однородны.

$H_1\colon$ Не однородны.

```python
scipy.stats.mannwhitneyu(data_0, data_1)
```

P-value малое, то гипотеза отвергается.
